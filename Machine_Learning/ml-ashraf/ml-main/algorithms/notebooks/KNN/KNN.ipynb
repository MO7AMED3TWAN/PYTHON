```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def KNN(xTrain, yTrain, xTest, yTest, predected):
    
    # Step 1 : Calculate distances
    distances = np.sqrt((xTrain - xTest) ** 2 + (yTrain - yTest) ** 2)
    
    # Step 2 : Merge distances and predected values
    dataLabel = pd.DataFrame(distances, columns=["Distances"])
    dataLabel["Pre"] = predected
    
    # Step 3 : Sort the data frame by distances
    dataLabel = (dataLabel.sort_values("Distances")).reset_index(drop=True)
    
    # Step 4 : Predect the input data
    if dataLabel["Pre"][0] == 0:
        out = 0
    else:
        out = 1
    
    # Step 5 : Plotting the data
    color = ["green" if i==1 else "red" for i in predected]
    plt.scatter(xTrain,yTrain, c=color)
    plt.scatter(xTest,yTest, color = 'blue')
    
    return out
```

```python
dataset = pd.read_csv("iphone_purchase_records.csv")
```

```python
x = int(input("Enter Age: "))
y = int(input("Enter Salary: "))


res = KNN(dataset["Age"], dataset["Salary"], x, y, dataset["Purchase Iphone"])

if res == 1:
    print("He will buy an iphon!")
else:
    print("He will not by an iphone!")
```
